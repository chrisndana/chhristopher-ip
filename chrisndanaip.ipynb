```python
#  DATA CLEANING
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# loading the file from desktop and creating a dataftame
stockdata=pd.read_csv("archive(1).zip",encoding='latin1')
df=pd.DataFrame(stockdata)
df
```

```python
# finding the sum of all null values in the data
df.isnull().sum()
```

```python
# dropping all the null values 
df1=df.dropna()
df1
```

```python
# findimg the shape to see if the null values have been dropped
df1.shape
```

```python
# finding duplicates within my data and dropping them 
df1.duplicated().sum()
df2=df1.drop_duplicates()
df2
```

```python
# finding the upper limit that will help me in dropping outliers 
upper_limit=df3.UnitPrice.mean()+3*df3.UnitPrice.std()
upper_limit
```

```python
# finding the lower limit 
lower_limit=df3.UnitPrice.mean()-3*df3.UnitPrice.std()
lower_limit
```

```python
df3[(df3.UnitPrice>upper_limit)| (df3.UnitPrice<lower_limit)]
```
